```python
#Then Sort the values on the dataframe using the Social Media Platform using the following function to confirm if row 0 is indeed removed: 

df.sort_values('Social Media Platform')
```

```python
#Rename the "Number of users" column to its specific country using the following function: 

df.columns = df.columns.str.replace('Number of users', 'United States')
print(df)
```

```python
#Answering first question: 

#1: When I choose a social media network, I am able to get a view the number of users in each country.
```

```python
#I have sorted the values in descending order. 

df.sort_values('Social Media Platform', ascending=False)
```

```python
#Answering the second question. 
#2: When I choose a country, I am able to see all the relevant social networks its people use ranked in descending order. 
```

```python
#Viewing the country and the relevant social networks its people use ranked in descending order using the function referenced below.

df1 = df.loc[:,["Social Media Platform", "China"]]
print(df1)
```

```python
# Ranked the Social Media Platforms in China in descending order.

df1.sort_values(by=['China','Social Media Platform'], inplace=True,
               ascending = [False, True])
print(df1)
```
